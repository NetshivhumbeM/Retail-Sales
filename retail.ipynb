```python
import pandas as pd
```

```python

import os
pwd = os.getcwd()
data = pd.read_csv(pwd+"/retail_sales_dataset.csv")
data
```

```python
data_set =data["Date"].str.split('-',expand=True)
data_set.columns =['Year','Month','Day']
data_set
```

Join the date table to the original data set

```python
data = pd.concat([data,data_set],axis=1)
## Drop the orignal date column (Not needed any more)
data.drop('Date',axis=1,inplace=True)
data
```

```python
## I want to know if we are only dealing with 2023 data so i can collapse the year column
no_years =data['Year'].nunique()
uni_years =data['Year'].unique()
print("This data is for",no_years,"distinct year/years which is/are",uni_years)
```

```python
uni_years =data['Year'].unique()
uni_years
```

```python
##sales trends over years and months
sales_years = data.groupby('Year')['Total Amount'].sum().reset_index()
sales_years=sales_years.sort_values(by="Total Amount",
                     axis=0,
                     ascending= False)
sales_years
```

```python
import calendar
data['Month'] =data['Month'].astype(int)
def month_number_to_name(Month):
    if 1 <= Month <= 12:
        return calendar.month_name[Month]
data['Month name'] = data['Month'].apply(month_number_to_name)
data.drop('Month',axis=1,inplace=True)
data
```

```python
sales_months = data.groupby('Month name')['Total Amount'].sum().reset_index()
sales_months=sales_months.sort_values(by="Total Amount",
                     axis=0,
                     ascending= False)
sales_months
```

```python
sales_day = data.groupby('Day')['Total Amount'].sum().reset_index()
sales_day=sales_day.sort_values(by="Total Amount",
                     axis=0,
                     ascending= False)
sales_day
```

```python
## Distribution of custormers by gender and age
cus_dis_gender=data.groupby('Gender')['Customer ID'].count().reset_index()
cus_dis_gender
```

```python
data['Age']= data['Age'].astype(int)
def no_customer_age(data,age_max,age_min):
    return ((age_min <= data['Age']) & (data['Age']<= age_max)).sum()

##The code ask the person between which ages does he or she wants to find out the number of custormers within that age
print("Age boundries to determine the number of customers withn those age")
age_min =int(input("Enter the minimum age: "))
age_max =int(input("Enter the maximum age: "))


number_customers =no_customer_age(data,age_max=age_max,age_min=age_min)
print("The number of customers between the age of", age_min,"years and",age_max,"years are",number_customers)


```

Top Selling Products

```python
products_quant = data.groupby('Product Category')['Quantity'].sum().reset_index()

products_sales = data.groupby('Product Category')['Total Amount'].sum().reset_index()

products_tab =pd.merge(products_quant,products_sales,on="Product Category",how='inner')
products_tab=products_tab.sort_values(by="Total Amount",
                     axis=0,
                     ascending= False)
products_tab
```

Purchasing behaviour for customers

```python
average_purchase_amount = data['Total Amount'].mean()
median_purchase_amount = data['Total Amount'].median()
min_purchase_amount = data['Total Amount'].min()
max_purchase_amount = data['Total Amount'].max()

print("The average amount a customer uses is",average_purchase_amount,"whereas the median is",median_purchase_amount,
      "with the maximum and minimum being",max_purchase_amount,"and",min_purchase_amount,"respectively ")
```

Product Pricing Analysis

```python
product_price_analysis = data[['Product Category','Price per Unit']]
ave_price_unit_category= product_price_analysis.groupby('Product Category')['Price per Unit'].mean().reset_index()

ave_price_unit_category.rename(columns={'Price per Unit':'Average Per price unit'},inplace=True)
ave_price_unit_category = ave_price_unit_category.sort_values(by= "Average Per price unit",
                                                              axis=0,
                                                              ascending= False)
ave_price_unit_category
```

Sales by gender

```python
sales_gender = data.groupby("Gender")['Total Amount'].sum().reset_index()
sales_gender = sales_gender.sort_values(by='Total Amount',
                                        axis=0,
                                        ascending=True)
sales_gender['difference']=sales_gender['Total Amount'].diff()
sales_gender
```

```python
differnc = sales_gender.loc[0,'difference']
print("The difference between the sales by gender is",differnc)
```

Age based purchasing patterns

```python
print("Age boundries to determine how purchasing patterns differ among different age group")
min_age =int(input("Enter the minimum age: "))
max_age =int(input("Enter the maximum age: "))
by_age_group =data[((min_age <= data['Age']) & (data['Age']<= max_age))]

sales_by_age_group =by_age_group['Total Amount'].sum()
print("The total amount generated among",min_age,"years and",max_age,"years age group is",sales_by_age_group)
```
